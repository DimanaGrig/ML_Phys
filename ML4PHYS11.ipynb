# applying ReLu instead of Sigmoid

```python
N0=2
N1=4
N2=1
```

```python
def apply_layer_new_relu(y_in,w,b): 
    z=dot(y_in,w)+b 
    return(z*(z>0))
```

```python
#list[:,n] selects all the entries that are in the n-th column of the vectors of the numpy array.
```

```python
def apply_multi_net_relu(y_in):
    global w, b, w_in, b_in, w_out, b_out, NLayers
    y=apply_layer_new_relu(y_in,w_in,b_in)  
    for j in range(NLayers):
        y=apply_layer_new_relu(y,w[j,:,:],b[j,:])
    output=apply_layer_new_relu(y,w_out, b_out)
    return(output)

```

```python
NLayers=20
LayerSize=100

w=random.uniform(low=-10, high=10, size =[NLayers, LayerSize, LayerSize])
b=random.uniform(low=-1, high=1, size=[NLayers, LayerSize])

w_in=random.uniform(low=-10, high=10, size =[2, LayerSize])
b_in=random.uniform(low=-1, high=1, size=LayerSize)

w_out=random.uniform(low=-10, high=10, size =[LayerSize,1])
b_out=random.uniform(low=-1, high=1, size=1)
```

```python
M=400
v0,v1=meshgrid(linspace(-0.5,0.5,M),linspace(-0.5,0.5,M))
batchsize=M**2 
y_in=zeros([batchsize,2])
y_in[:,0]=v0.flatten()
y_in[:,1]=v1.flatten()
```

```python
y_out=apply_multi_net_relu(y_in)
```

```python
y_2D=reshape(y_out[:,0],[M,M]) 
plt.figure(figsize=[5,5])
plt.axes([0,0,1,1]) 
plt.imshow(y_2D,origin='lower',extent=[-0.5,0.5,-0.5,0.5],interpolation='nearest')
plt.show()
```
